```python
path = 'C:\\Users\\kayde\\Desktop\\ENGL 343\\Verne Texts\\'

import glob
files = glob.glob(path + "*.txt")
```

```python
import pandas as pd

texts = [f[48:-4] for f in files]
date = [f[44:48] for f in files]
df = pd.DataFrame(files, columns=['FILE_PATH'])
df['TEXT'] = texts
df['DATE'] = date
df
```

```python
import codecs
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize

df['FULL_TEXT'] = ''
df['FULL_TEXT_NOLOWER'] = ''

for index, row in df.iterrows():
    filepath = row['FILE_PATH']
    text = codecs.open(filepath, "r", encoding='utf8')         
    raw = text.read()
    raw1 = raw.lower()
    df.at[index, 'FULL_TEXT'] = raw1
    df.at[index, 'FULL_TEXT_NOLOWER'] = raw
df
```

```python
import re

def clean_string(s): #Cleans strings for the vectorizer

    cleaned = re.sub('\n', ' ', s) #Replaces newline characters with space
    cleaned = re.sub('(\w+)?@\w+',' ',cleaned) #Removes whole word if it has an '@' (i.e. usernames)
    cleaned_str=re.sub('(\s+)',' ',cleaned) #Remove multiple spaces
    cleaned_str=cleaned_str.lower() #Lower case

    return cleaned_str
```

```python
proper_df = df[df.DATE != '0000']
clean_proper_df = proper_df.copy()
clean_proper_df['FULL_TEXT'] = clean_proper_df['FULL_TEXT'].apply(clean_string)
clean_proper_df['FULL_TEXT_NOLOWER'] = clean_proper_df['FULL_TEXT_NOLOWER'].apply(clean_string)
```

```python
verne_list = []

for content in clean_proper_df['FULL_TEXT']:
    verne_list.append(content)
len(verne_list), len(verne_list[0])
```

```python
import nltk
nltk.download('punkt')
verne_sentences = [sentence for novel in verne_list for sentence in nltk.sent_tokenize(novel)]
```

```python
def fast_tokenize(text):
    from string import punctuation
    no_punct = "".join([char for char in text if char not in punctuation])    
    tokens = no_punct.split()
    
    return tokens
```

```python
words_by_verne_sentence = [fast_tokenize(sentence.lower()) for sentence in verne_sentences]
words_by_verne_sentence[0], len(words_by_verne_sentence)
```

```python
import gensim
model = gensim.models.Word2Vec(words_by_verne_sentence, window=5, \
                               min_count=5, sg=0, alpha=0.025, batch_words=10000)
```

```python
model.wv.most_similar(positive=['god'], topn=20)
```

```python
female_vector = model.wv.most_similar(positive=['female'], topn=20)

female_tokens = [token for token, weight in female_vector]
female_vectors = [model.wv[word] for word in female_tokens] 
```

```python
from sklearn.metrics import pairwise
dist_matrix = pairwise.pairwise_distances(female_vectors, metric='cosine')
```

```python
from sklearn.manifold import MDS
mds = MDS(n_components = 2, dissimilarity='precomputed')
embeddings = mds.fit_transform(dist_matrix)
```

```python
import matplotlib.pyplot as plt

_, ax = plt.subplots(figsize=(10,10))
ax.scatter(embeddings[:,0], embeddings[:,1], alpha=0)
for i in range(len(female_vectors)):
    ax.annotate(female_tokens[i], ((embeddings[i,0], embeddings[i,1])))
```
